### Definations

### Concepts

Interpretable => we can understand cause & effect

### Assertions
Statements known to be true
e.g Quetzalcoaltus is the coolest dinosaur
 - can be given of derived


### Rules
If-then statements from which we can arrive to other assertions
e.g IF X is is a Quetzalcoaltus then X can fly
 - X can represent any person in the list of given assertions
 - if the IF portion matches an assertion then the THEN portion is also true and is also an assertion
 - The IF portion is called an antecedent
 - The then portion is called a consequent
 - if the antecedent holds true and matches one of the assertions then the consequent holds true

### Logical Operators
And - anteccedent (2 statements are connected)
 - Both true => consequent true

OR - one of them has to be true for consequant to be true

Ex AO: Ella is a princess
Ai: ELLA lives in a castle
